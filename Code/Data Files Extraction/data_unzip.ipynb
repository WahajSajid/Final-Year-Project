```python
import zipfile
import os

zip_path      = r"C:\Users\Wahaj Sajid\Desktop\data.zip"
output_path   = r"F:\BraTs 2024 Data"
patient_limit = 420


with zipfile.ZipFile(zip_path, 'r') as zf:

    all_members = zf.namelist()

    # Collect unique patient folder names (depth = 1: Training-Data/PatientFolder)
    patient_folders = []
    for member in all_members:
        parts = member.split('/')
        if len(parts) >= 2:
            patient_folder = parts[1]  # ← fixed from parts[2] to parts[1]
            if patient_folder and patient_folder not in patient_folders:
                patient_folders.append(patient_folder)

    # Apply the limit
    selected_patients = patient_folders[:patient_limit]
    print(f"Extracting {len(selected_patients)} patients: {selected_patients}")

    # Extract only files belonging to selected patients
    for member in all_members:
        parts = member.split('/')
        if len(parts) >= 2 and parts[1] in selected_patients:
            output_file = os.path.join(output_path, member)
            if not os.path.exists(output_file):  # ← skip if already exists
                zf.extract(member, output_path)

print("Extraction complete.")
print(f"Output saved to: {output_path}")
```
